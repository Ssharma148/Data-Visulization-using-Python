### 1. Compare Sales by region for 2016 with 2015 using bar chart

### 6. Create a new column Qtr in the imported dataset. The Quarters are based on months and defined as -
#### â€¢ Jan - Mar : Q1
#### â€¢ Apr - Jun : Q2
#### â€¢ Jul - Sep : Q3
#### â€¢ Oct - Dec : Q4

```python
month = sales["Month"]
quarter = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
sales["Qtr"]= pd.Series(quarter)
```

### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot

```python
QtrWise_sales = sales.groupby("Qtr")["Sales2015","Sales2016"].sum()
QtrWise_sales
```

```python
QtrWise_sales.plot(kind="bar")
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```
